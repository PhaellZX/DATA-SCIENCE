```python
import pandas as pd
base = pd.read_csv('supermarket_sale.csv')
base.head()
```

```python
base.tail()
```

```python
base = base.rename({'Product line':'Product_line'}, axis = 1)
base['Product_line'].value_counts()
```

## Product Line
0 - Fashion accessories

1 - Food and beverages

2 - Electronic accessories

3 - Sports and travel

4 - Home and lifestyle

5 - Health and beauty

```python
base['Product_line'] = base['Product_line'].replace('Fashion accessories', 0)
base['Product_line'] = base['Product_line'].replace('Food and beverages', 1)
base['Product_line'] = base['Product_line'].replace('Electronic accessories', 2)
base['Product_line'] = base['Product_line'].replace('Sports and travel', 3)
base['Product_line'] = base['Product_line'].replace('Home and lifestyle', 4)
base['Product_line'] = base['Product_line'].replace('Health and beauty', 5)
base.head()
```

## Branch
0 - Yangon(A)

1 - Mandalay(B)

2 - Naypyitaw(C)

```python
base['Branch'] = base['Branch'].replace('A', 0)
base['Branch'] = base['Branch'].replace('B', 1)
base['Branch'] = base['Branch'].replace('C', 2)
base.head()
```

```python
base = base.rename({'Customer type':'Customer_type'}, axis = 1)
base['Customer_type'].value_counts()
```

## Customer Type
0 - Member

1 - Normal

```python
base['Customer_type'] = base['Customer_type'].replace('Member', 0)
base['Customer_type'] = base['Customer_type'].replace('Normal', 1)
base.head()
```

## Gender
0 - Female

1 - Male

```python
base['Gender'] = base['Gender'].replace('Female', 0)
base['Gender'] = base['Gender'].replace('Male', 1)
base.head()
```

```python
base['Payment'].value_counts()
```

## Payment
0 - Ewallet

1 - Cash

2 - Credit card

```python
base['Payment'] = base['Payment'].replace('Ewallet', 0)
base['Payment'] = base['Payment'].replace('Cash', 1)
base['Payment'] = base['Payment'].replace('Credit card', 2)
base.head()
```

## Deleting object columns

```python
columns = base.dtypes[base.dtypes.values == 'object'].index
```

```python
base = base.drop(columns, axis = 1)
base.info()
```

```python
base = base.astype(int).astype(float)
base.head(10)
```

```python
base.loc[(base.Rating >= 7.0),'product_is_good'] = 1.0
base.head()
```

```python
base.loc[(base.Rating < 7.0),'product_is_good'] = 0.0
base.head()
```

## Making the last changes to the Dataset

```python
base = base.rename({'Unit price':'Unit_price'}, axis = 1)
base = base.rename({'Tax 5%':'Tax_5%'}, axis = 1)
base = base.rename({'gross margin percentage':'gross_margin_percentage'}, axis = 1)
base = base.rename({'gross income':'gross_income'}, axis = 1)
base.head()
```

## Training and Testing the model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.linear_model import LogisticRegression
```

```python
teste = pd.read_csv('supermarket_sale.csv')
teste = base
teste.head()
```

```python
X_train = base.drop('product_is_good',axis=1)
y_train = base.product_is_good
```

```python
X_test = base.drop('product_is_good',axis=1)
y_test = base.product_is_good
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
neight = neigh.fit(X_train, y_train)
clfArvore = tree.DecisionTreeClassifier(random_state=0)
clfArvore = clfArvore.fit(X_train, y_train)
clfLog = LogisticRegression(random_state=0,max_iter=1000).fit(X_train, y_train)
clfLog = clfLog.fit(X_train, y_train)
```

```python
pred_KNN = neigh.predict(X_test)
pred_Arvore = clfArvore.predict(X_test)
pred_Log = clfLog.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
# Accuracy
print(accuracy_score(y_test,pred_KNN))
print(accuracy_score(y_test,pred_Arvore))
print(accuracy_score(y_test,pred_Log))   
```

```python
# Precision
print(precision_score(y_test,pred_KNN))
print(precision_score(y_test,pred_Arvore))
print(precision_score(y_test,pred_Log)) 
```

```python
# Recall
print(recall_score(y_test,pred_KNN))
print(recall_score(y_test,pred_Arvore))
print(recall_score(y_test,pred_Log)) 
```
