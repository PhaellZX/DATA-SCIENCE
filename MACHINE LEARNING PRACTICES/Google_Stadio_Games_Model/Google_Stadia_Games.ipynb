## Importando o Dataset 

```python
import pandas as pd
```

```python
base = pd.read_csv('df_1.csv')
```

```python
base.tail()
```

## Mudando o nomes das colunas

```python
# Changing the name of the 4 columns
base = base.rename({'Stadia release date':'Stadia_release_date'}, axis = 1)
base = base.rename({'Original release date[a]':'Original_release_date'}, axis = 1)
base = base.rename({'Date added to Stadia Pro[b]':'Date_added_Stadia_Pro'}, axis = 1)
base = base.rename({'Date no longer claimable on Stadia Pro[c]':'Date_Stadia_Pro'}, axis = 1)
base.head(3)
```

## Removendo o dia e o mês e colocando somente o ano nas colunas Stadia_release_date	Original_release_date	Date_added_Stadia_Pro	Date_Stadia_Pro

```python
# Stadia_release_date
i = 0
for i in range(0, 275):
    value = base.Stadia_release_date[i]
    mes, num, ano = value.split(" ")
    base.Stadia_release_date[i] = ano
    i += 0
```

```python
base.head()
```

```python
base.tail()
```

```python
# Original_release_date
i = 0
for i in range(0, 275):
    value = base.Original_release_date[i]
    mes, num, ano = value.split(" ")
    base.Original_release_date[i] = ano
    i += 0
```

Obs: Está dando erro nessa parte... tive que fazer o mesmo processo da linha 58 em diante

```python
i = 58
for i in range (58, 275):
    value = base.Original_release_date[i]
    mes, ano = value.split(", ")
    base.Original_release_date[i] = ano
    i += 1
```

```python
base.head()
```

```python
base.tail()
```

## Substituindo os valores 'NaN' para '0' 

```python
base['Date_added_Stadia_Pro'] = base['Date_added_Stadia_Pro'].replace(float('NaN'),'0')
base['Date_Stadia_Pro'] = base['Date_Stadia_Pro'].replace(float('NaN'),'0')
base
```

## Continuando  a adicionar os anos nas colunas

```python
# Date_added_Stadia_Pro
i = 0
for i in range(0, 275):
    if base.Date_added_Stadia_Pro[i] == '0':
        base.Date_added_Stadia_Pro[i] = '0'
    else:
        value = base.Date_added_Stadia_Pro[i]
        mes, ano = value.split(", ")
        base.Date_added_Stadia_Pro[i] = ano
    i += 1
```

```python
# Date_Stadia_Pro
i = 0
for i in range(0, 275):
    if base.Date_Stadia_Pro[i] == '0':
        base.Date_Stadia_Pro[i] = '0'
    else:
        value = base.Date_Stadia_Pro[i]
        mes, ano = value.split(", ")
        base.Date_Stadia_Pro[i] = ano
    i += 1
```

```python
base.head()
```

```python
base.tail()
```

## Mudando o nome da coluna Unnamed: 0 e Genre(s)

```python
base = base.rename({'Genre(s)':'Genre'}, axis = 1)
base = base.rename({'Unnamed: 0':'Id'}, axis = 1)
base.head()
```

## Removendo Colunas que não serão utilizadas

```python
base = base.drop('Ref.',axis=1)
base = base.drop('Title',axis=1)
base = base.drop('Genre',axis=1)
base = base.drop('Developer(s)',axis=1)
base = base.drop('Publisher(s)',axis=1)
base.head()
```

```python
base.Date_Stadia_Pro[224] = '2020'
base.Date_Stadia_Pro[42] = '2020'
```

Obs: No momento que ia converter os valores strings para inteiros, apareceu 2 dados com anos '2020[a]' tive que procurar entre a colunas e as linhas esse dados.

## Convertendo Todos os Dados para inteiros

```python
base = base.astype(str).astype(int)
base
```

## Mostrando os dados convertidos

```python
base.info()
```

## Criando uma nova coluna como saída
se data do game stadia pro for acima ou igual a 2021 é lançamento 

```python
base.loc[base['Date_Stadia_Pro'] > 2021,'lancamento'] = 1
base.loc[base['Date_Stadia_Pro'] <= 2021,'lancamento'] = 0
base
```

## Modelo pronto para treinamento

```python
base
```

# Treinando o Modelo

```python
X = base.drop('lancamento',axis=1)
y = base.lancamento
```

## KNN -  KNeighborsClassifier

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
```

```python
neigh = KNeighborsClassifier(n_neighbors=3)
```

```python
neigh.fit(X, y)
```

```python
neigh.score(X, y)
```

## Arvore de Decisão

```python
# Desicion Tree
from sklearn import tree
```

```python
clfArvore = tree.DecisionTreeClassifier(random_state=0)
```

```python
clfArvore = clfArvore.fit(X, y)
```

```python
clfArvore.score(X,y)
```

## Regressão Linear

```python
# Linear Regression
from sklearn.linear_model import LogisticRegression
```

```python
clfLog = LogisticRegression(random_state=0,max_iter=1000).fit(X, y)
```

```python
clfLog.score(X,y)
```

## Criando a base de teste

```python
teste = pd.read_csv('df_1.csv')
teste = base
teste
```

# Avaliando o Modelo

```python
X_treino = base.drop('lancamento',axis=1)
y_treino = base.lancamento
```

```python
X_teste = teste.drop('lancamento',axis=1)
y_teste = teste.lancamento
```

```python
neight = neigh.fit(X_treino, y_treino)
clfArvore = clfArvore.fit(X_treino, y_treino)
clfLog = clfLog.fit(X_treino, y_treino)
```

```python
pred_KNN = neigh.predict(X_teste)
pred_Arvore = clfArvore.predict(X_teste)
pred_Log = clfLog.predict(X_teste)
```

```python
# Importando as métricas de avaliação
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
# Acurácia
print(accuracy_score(y_teste,pred_KNN))
print(accuracy_score(y_teste,pred_Arvore))
print(accuracy_score(y_teste,pred_Log))    
```

```python
# Precisão
print(precision_score(y_teste,pred_KNN))
print(precision_score(y_teste,pred_Arvore))
print(precision_score(y_teste,pred_Log)) 
```

```python
# Recall
print(recall_score(y_teste,pred_KNN))
print(recall_score(y_teste,pred_Arvore))
print(recall_score(y_teste,pred_Log)) 
```

Obs: Arvore de Decisão teve o melhor desempenho
