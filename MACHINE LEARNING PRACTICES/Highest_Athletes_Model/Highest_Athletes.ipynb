```python
import pandas as pd
data = pd.read_csv('highest.csv')
data.head()
```

```python
data.tail()
```

## Making the last changes to the Dataset

```python
data = data.rename({'RANK':'rank'}, axis = 1)
data = data.rename({'NAME':'name'}, axis = 1)
data = data.rename({' SPORT':'sport'}, axis = 1)
data = data.rename({' TOTAL EARNINGS':'total_earnings'}, axis = 1)
data = data.rename({' ON-THE-FIELD EARNINGS':'on_the_field_earnings'}, axis = 1)
data = data.rename({'OFF-THE-FIELD EARNINGS':'off_the_field_earnings'}, axis = 1)
data.head()
```

```python
data['sport'].value_counts()
```
