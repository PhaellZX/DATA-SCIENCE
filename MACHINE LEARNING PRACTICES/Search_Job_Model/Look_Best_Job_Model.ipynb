```python
import pandas as pd
```

```python
base = pd.read_csv('salaries.csv')
```

```python
base.head(10)
```

```python
base.info()
base.shape
```

```python
base.experience_level.value_counts()
```

```python
'''
SE Senior-level / Expert -> 1
MI Mid-level / Intermediate -> 2
EN Entry-level / Junior -> 3
EX Executive-level / Director -> 4
'''
base['experience_level'] = base['experience_level'].replace('SE',1)
base['experience_level'] = base['experience_level'].replace('MI',2)
base['experience_level'] = base['experience_level'].replace('EN',3)
base['experience_level'] = base['experience_level'].replace('EX',4)

base.head(10)
```

```python
base.employment_type.value_counts()
```

```python
'''
PT: Part-time -> 1
FT: Full-time -> 2
CT: Contract -> 3
FL: Freelance -> 4
'''

base['employment_type'] = base['employment_type'].replace('PT',1)
base['employment_type'] = base['employment_type'].replace('FT',2)
base['employment_type'] = base['employment_type'].replace('CT',3)
base['employment_type'] = base['employment_type'].replace('FL',4)

base.head(10)
```

```python
base = base.drop('job_title',axis=1)
base.head(10)
```

```python
base.salary_currency.value_counts()
```

```python
'''
USD -> 10
EUR -> 20
GBP -> 30
INR -> 40
CAD -> 50
AUD -> 60
BRL -> 70
PLN -> 80
TRY -> 90
JPY -> 100     
SGD -> 110
DKK -> 120
CNY -> 130
HUF -> 140
MXN -> 150
CHF -> 160      
CZK -> 170
CLP -> 180     
'''

base['salary_currency'] = base['salary_currency'].replace('USD',10)
base['salary_currency'] = base['salary_currency'].replace('EUR',20)
base['salary_currency'] = base['salary_currency'].replace('GBP',30)
base['salary_currency'] = base['salary_currency'].replace('INR',40)
base['salary_currency'] = base['salary_currency'].replace('CAD',50)
base['salary_currency'] = base['salary_currency'].replace('AUD',60)
base['salary_currency'] = base['salary_currency'].replace('BRL',70)
base['salary_currency'] = base['salary_currency'].replace('PLN',80)
base['salary_currency'] = base['salary_currency'].replace('TRY',90)
base['salary_currency'] = base['salary_currency'].replace('JPY',100)
base['salary_currency'] = base['salary_currency'].replace('SGD',110)
base['salary_currency'] = base['salary_currency'].replace('DKK',120)
base['salary_currency'] = base['salary_currency'].replace('CNY',130)
base['salary_currency'] = base['salary_currency'].replace('HUF',140)
base['salary_currency'] = base['salary_currency'].replace('MXN',150)
base['salary_currency'] = base['salary_currency'].replace('CHF',160)
base['salary_currency'] = base['salary_currency'].replace('CZK',170)
base['salary_currency'] = base['salary_currency'].replace('CLP',180)

base.head(10)
```

```python
base = base.drop('employee_residence', axis=1)
base.head(10)
```

```python
base = base.drop('company_location',axis=1)
base.head(10)
```

```python
base.company_size.value_counts()
```

```python
'''
S: less than 50 employees (small) -> 0
M: 50 to 250 employees (medium) -> 1
L: more than 250 employees (large) -> 2
'''

base['company_size'] = base['company_size'].replace('S',0)
base['company_size'] = base['company_size'].replace('M',1)
base['company_size'] = base['company_size'].replace('L',2)

base.head(10)
```

```python
base = base.drop('remote_ratio',axis=1)
base.head(10)
```

```python
base.info()
```

```python
base.loc[(base['salary'] <= 150000) & (base['salary_currency'] != 10) | (base['salary_currency'] != 20),'best_indication'] = 0
base
```

```python
base.loc[(base['salary'] >= 150000) & (base['salary_currency'] == 10) | (base['salary_currency'] == 20),'best_indication'] = 1
base
```

```python
base['best_indication'].describe()
```

```python
import matplotlib.pyplot as plt
```

```python
x = base.best_indication

# plot
fig, ax = plt.subplots()

ax.hist(x, bins=20, linewidth=0.5, edgecolor="white", label='Best Indication') 
ax.legend()

plt.show()
```

```python
teste = pd.read_csv('salaries.csv')
teste = base
teste.head(10)
```

## Training the model

```python
X = base.drop('best_indication',axis=1)
y = base.best_indication
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
neigh.score(X, y)
```

```python
# Decisiom Tree
from sklearn import tree
clfArvore = tree.DecisionTreeClassifier(random_state=0)
clfArvore = clfArvore.fit(X, y)
clfArvore.score(X,y)
```

```python
# Linear Regression
from sklearn.linear_model import LogisticRegression
clfLog = LogisticRegression(random_state=0,max_iter=1000).fit(X, y)
clfLog.score(X,y)
```

## Evaluating the model 

```python
X_treino = base.drop('best_indication',axis=1)
y_treino = base.best_indication

X_teste = teste.drop('best_indication',axis=1)
y_teste = teste.best_indication
```

```python
# KNN
neight = neigh.fit(X_treino, y_treino)
clfArvore = clfArvore.fit(X_treino, y_treino)
clfLog = clfLog.fit(X_treino, y_treino)
```

```python
pred_KNN = neigh.predict(X_teste)
pred_Arvore = clfArvore.predict(X_teste)
pred_Log = clfLog.predict(X_teste)
```

```python
# Importando as métricas de avaliação
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
# Acurácia
print(accuracy_score(y_teste,pred_KNN))
print(accuracy_score(y_teste,pred_Arvore))
print(accuracy_score(y_teste,pred_Log))    
```

```python
# Precisão
print(precision_score(y_teste,pred_KNN))
print(precision_score(y_teste,pred_Arvore))
print(precision_score(y_teste,pred_Log)) 
```

```python
# Recall
print(recall_score(y_teste,pred_KNN))
print(recall_score(y_teste,pred_Arvore))
print(recall_score(y_teste,pred_Log)) 
```
